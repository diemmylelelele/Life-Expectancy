```python
# Import statements
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# Upload dataset
data=pd.read_csv('/content/life expectancy.csv')
```

```python
# Create figure
plt.figure(figsize=(8, 4))
# Plot histogram
plt.hist(data['Life_expectancy'], bins=25, edgecolor='white', color='blue')
# Add title and labels for x,y-axis
plt.title('Histogram of Life Expectancy Globally', fontsize=13, fontfamily='serif')
plt.xlabel('Life Expectancy', fontsize=10, fontweight='normal', fontfamily='sans-serif')
plt.ylabel('Frequency', fontsize=10, fontweight='normal', fontfamily='sans-serif')
# Add grid
plt.grid(True)
# Show the plot
plt.show()
```

```python
# Group the data by region and year, and calculate the average life expectancy
life_expectancy_by_region = data.groupby(['Region', 'Year'])['Life_expectancy'].mean().reset_index()
# Create figure
plt.figure(figsize=(18, 10))
# Iterate through each unique region in the dataset
for region in life_expectancy_by_region['Region'].unique():
    region_data = life_expectancy_by_region[life_expectancy_by_region['Region'] == region]
    plt.plot(region_data['Year'], region_data['Life_expectancy'], marker='o', linestyle='-', label=region)
# Add title and label for x,y-axis
plt.title('Life Expectancy by Region 2000 - 2015', fontsize=13, fontfamily='serif' )
plt.xlabel('Year',fontsize=10, fontweight='normal', fontfamily='sans-serif')
plt.ylabel('Average Life Expectancy', fontsize=10, fontweight='normal', fontfamily='sans-serif' )
# Add grid
plt.grid(True)
# Add legend
plt.legend()
# Show the plot
plt.show()
```

```python
# create a figure
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Life_expectancy', data=data, palette='Paired', flierprops={'marker': 'o', 'markerfacecolor': 'red'})
# Add title and labels for x, y axis
plt.title('Box Plot of Life Expectancy by Region', fontsize=13, fontfamily='serif')
plt.xlabel('Region', fontsize=10, fontweight='normal', fontfamily='sans-serif')
plt.ylabel('Life Expectancy', fontsize=10, fontweight='normal', fontfamily='sans-serif')
# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)
# Show the plot
plt.show()
```

```python
correlation_matrix = data.corr()
life_expectancy_correlation = correlation_matrix['Life_expectancy']
# Sort the correlation coefficient with life expectancy
correlations = life_expectancy_correlation.sort_values(ascending=False)
correlations
```

```python
# Plot the heatmap of correlation of variables
plt.figure(figsize=(8, 4))
sns.heatmap(data.corr(), cmap='RdBu')
plt.title('Correlation Heatmap', fontsize=13, fontfamily='serif')
```

```python
# Define the factors and their correlation coefficients
positive_factors = ['Schooling', 'Polio', 'Diphtheria','GDP_per_capita' ]
# Replace with actual correlation coefficients
correlations = [0.732484, 0.641217, 0.627541, 0.583090]
# Create a subplot for each scatter plot
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()
for i, factor in enumerate(positive_factors):
    ax = axes[i]
    # Create a scatter plot
    ax.scatter(data[factor], data['Life_expectancy'], alpha=0.5, color='blue')
    ax.set_title(f'{factor} vs Life Expectancy', fontsize=10)
    ax.set_xlabel(factor, fontsize=10, fontweight='normal', fontfamily='sans-serif')
    ax.set_ylabel('Life Expectancy', fontsize=10, fontweight='normal', fontfamily='sans-serif')
    ax.grid(True)
    # Add correlation coefficient to the plot
    ax.text(0.05, 0.95, f'Correlation: {correlations[i]:.2f}', transform=ax.transAxes, fontsize=10, verticalalignment='top', color='red')
# Adjust layout
plt.tight_layout()
# Show the figure
plt.show()
```

```python
# Group the data by region and calculate the mean for each variable
region_data = data.groupby('Region')[['Schooling', 'GDP_per_capita', 'Polio', 'Diphtheria']].mean()
# Create a figure with four horizontal bar charts
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.ravel()
# Define colors for bars
colors = ['skyblue', 'lightcoral', 'lightgreen', 'lightsalmon']
# Create horizontal bar charts for Schooling, GDP, Polio, and Diphtheria
for i, var in enumerate(region_data.columns):
    ax = axes[i]
    ax.barh(region_data.index, region_data[var], color=colors[i])
    ax.set_xlabel(var)
    ax.set_title(f'{var} by Region')
# Adjust layout
plt.tight_layout()
# Show the combined figure
plt.show()
```

```python
# Define the factors and their correlation coefficients
negative_factors = ['Adult_mortality', 'Under_five_deaths','Incidents_HIV', 'Thinness_ten_nineteen_years',]
# Replace with actual correlation coefficients
correlations = [-0.945360, -0.920419,-0.920032,-0.467824]
# Create a subplot for each scatter plot
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()
for i, factor in enumerate(negative_factors):
    ax = axes[i]
    # Create a scatter plot
    ax.scatter(data[factor], data['Life_expectancy'], alpha=0.5, color='blue')
    ax.set_title(f'{factor} vs Life Expectancy')
    ax.set_xlabel(factor)
    ax.set_ylabel('Life Expectancy')
    ax.grid(True)
    # Add correlation coefficient to the plot
    ax.text(0.85, 0.92, f'Correlation: {correlations[i]:.2f}', transform=ax.transAxes,
            fontsize=10, verticalalignment='top', horizontalalignment='right', color='red')
# Adjust layout
plt.tight_layout()
# Show the figure
plt.show()
```

```python
# Replace values in the 'economy_status_Developed' and 'economy_status_Developing' columns
data['Economy_status_Developed'] = data['Economy_status_Developed'].replace({1: 'Developed', 0: 'Developing'})
data['Economy_status_Developing'] = data['Economy_status_Developing'].replace({1: 'Developed', 0: 'Developing'})
fig=plt.figure(figsize=(8,4))
# Plot the violin
sns.violinplot(data=data, x=data["Economy_status_Developed"], y=data["Life_expectancy"], hue=None ,color='blue')
# Add title and label for x,y axis
plt.title('Life expectancy vs Country Status', fontsize=12, fontfamily='serif')
plt.xlabel('Country status', fontsize=10, fontweight='normal', fontfamily='sans-serif')
plt.ylabel('Life Expectancy', fontsize=10, fontweight='normal', fontfamily='sans-serif')
# Show the plot
plt.show()
```
